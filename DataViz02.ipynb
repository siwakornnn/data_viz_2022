## Import Data

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df.head()
```

# Scatter Plot

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter([1,-1],[2,-1])
```

```python
plt.scatter(df['PetalLength'],df['SepalLength'])
```

## เปลี่ยนสี

```python
plt.scatter(df['PetalLength'],df['SepalLength'],c='r')
```

## ระบุสีให้แต่ละจุด

```python
set(df['Name'])
```

## แทนค่าดอกไม้แทนตัวอักษร

```python
df2 = df.replace({'Iris-setosa': 'r', 'Iris-versicolor' : 'g' ,'Iris-virginica' : 'b'})
```

```python
df2.head()
```

```python
plt.scatter(df['PetalLength'],df['SepalLength'],c=df2['Name'])
```

## เพิ่มรายละเอียดกราฟ

### ชื่อแกน

```python
plt.scatter(df['PetalLength'],df['SepalLength'],c=df2['Name'])
plt.xlabel('PetalLength')
plt.ylabel('SepalLength')
```

### ชื่อ marker

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50],label='setosa')
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100],label='versicolor')
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],c=df2['Name'][100:],label='virginica')
plt.legend()
plt.xlabel('PetalLength')
plt.ylabel('SepalLength')
```

### ชื่อกราฟ

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50],label='setosa')
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100],label='versicolor')
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],c=df2['Name'][100:],label='virginica')
plt.legend()
plt.xlabel('PetalLength')
plt.ylabel('SepalLength')
plt.title('comparing prtal&sepal length of differant iris flowers');
```

ใช้ขนาดของ marker ในการแสดงค่าใน column

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50],label='setosa',s=df2['PetalWidth'][:50])
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100],label='versicolor',s=df2['PetalWidth'][:50])
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],c=df2['Name'][100:],label='virginica',s=df2['PetalWidth'][100:])
plt.legend()
plt.xlabel('PetalLength')
plt.ylabel('SepalLength')
plt.title('comparing prtal&sepal length of differant iris flowers');
```

ขยายขนาด marker ที่เล็กเกินไป

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50],label='setosa',s=20*df2['PetalWidth'][:50])
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100],label='versicolor',s=20*df2['PetalWidth'][:50])
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],c=df2['Name'][100:],label='virginica',s=20*df2['PetalWidth'][100:])
plt.legend()
plt.xlabel('PetalLength')
plt.ylabel('SepalLength')
plt.title('comparing prtal&sepal length of differant iris flowers');
```

## ปรับความโปร่งใสของ markerเพื่อให้มองเห็นจุดที่ถูกบัง (Bubble Chart) 

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],alpha=0.5,c=df2['Name'][:50],label='setosa',s=20*df2['PetalWidth'][:50])
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],alpha=0.5,c=df2['Name'][50:100],label='versicolor',s=20*df2['PetalWidth'][:50])
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],alpha=0.5,c=df2['Name'][100:],label='virginica',s=20*df2['PetalWidth'][100:])
plt.legend()
plt.xlabel('PetalLength')
plt.ylabel('SepalLength')
plt.title('comparing prtal&sepal length of differant iris flowers');
```

## เปลี่ยนรูปร่าง marker

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],alpha=0.5,c=df2['Name'][:50],label='setosa',s=20*df2['PetalWidth'][:50],marker="*")
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],alpha=0.5,c=df2['Name'][50:100],label='versicolor',s=20*df2['PetalWidth'][:50],marker="*")
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],alpha=0.5,c=df2['Name'][100:],label='virginica',s=20*df2['PetalWidth'][100:],marker="*")
plt.legend()
plt.xlabel('PetalLength')
plt.ylabel('SepalLength')
plt.title('comparing prtal&sepal length of differant iris flowers');
```

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],alpha=0.5,c=df2['Name'][:50],label='setosa',s=20*df2['PetalWidth'][:50],marker="$s$")
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],alpha=0.5,c=df2['Name'][50:100],label='versicolor',s=20*df2['PetalWidth'][:50],marker="$c$")
plt.scatter(df2['PetalLength'][100:],df2['SepalLength'][100:],alpha=0.5,c=df2['Name'][100:],label='virginica',s=20*df2['PetalWidth'][100:],marker="$g$")
plt.legend()
plt.xlabel('PetalLength')
plt.ylabel('SepalLength')
plt.title('comparing prtal&sepal length of differant iris flowers');
```
