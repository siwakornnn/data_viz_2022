**Quiz (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)**

1. คำนวณ อายุเฉลี่ย ของผู้ชาย ในข้อมูลทั้งหมด
2. คำนวณ อายุเฉลี่ย ของผู้ชาย ในจังหวัดขอนแก่น
3. หาจำนวนผู้ป่วยที่เป็นคน ที่ไม่ใช่คนไทย

````python
this_data = data_covid[['No.','sex','age','province_of_onset','risk']]
this_data
````

````python
men = this_data[this_data['sex']=='ชาย']      #ใช้ข้อมูลในตาราง this_data โดยที่มีเงื่อนไขว่าต้องเป็น ชาย เท่านั้น
men['age']  # แสดงค่าออกมา
````

````python
sum = 0
N = 0
for a in men['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยเพศชาย {sum/N}')   # แสดงค่าออกมา
````

````python
men_KK = men[men['province_of_onset']=='ขอนแก่น']   #กำหนดตัวแปร men_KK ให้เลือกตาราง man ที่ column province_of_onset ให้เป็น ขอนแก่น 

sum = 0
N = 0
for a in men_KK['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยเพศชาย ในจังหวัดขอนแก่น {sum/N}')   # แสดงค่าออกมา
````

````python
foreigner  = this_data[this_data['risk']=='คนต่างชาติเดินทางมาจากต่างประเทศ']
Num = foreigner['No.']
foreigner[['No.','risk']]
````

````python
sum = 0
N = 0
for a in Num:
    
    if a >0:
        sum += 1 
        N += 1

print(f'ผู้ป่วยที่ไม่ใช่คนไทย {sum} คน')
````

## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021 

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021

````python
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````
