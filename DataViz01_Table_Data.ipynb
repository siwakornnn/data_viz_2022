```python
import pandas as pd  ## เรียกใช้ pandas
```

```python
data= pd.read_excel('/content/drive/MyDrive/dataviz_2021_data/confirmed-cases.xlsx')  ## สร้างตัวแปร รับค่าข้อมูลที่อยู่ใน Google drive
data ## แสดงข้อมูล
```

```python
df1= data[['sex','age','risk']] ## เลือก column ต้องการ ได้แก่ sex age และ risk
df1
```

```python
set(df1['risk']) ## ดูค่า column risk ในตาราง df1
```

```python
df1_1 = df1[df1['risk']=='อยู่ระหว่างสอบสวน']   ## เลือกข้อมูล risk ในตาราง df1 ที่เป็น  อยู่ระหว่างสอบสวน
df1_1
```

```python
df1_2 = df1[df1['risk']=='สนามมวย']   ## เลือกข้อมูล risk ในตาราง df1 ที่เป็น  สนามมวย
df1_2
```

```python
df1_3 = df1[df1['risk']=='สัมผัสผู้ติดเชื้อ'] ## เลือกข้อมูล risk ในตาราง df1 ที่เป็น สัมผัสผู้ติดเชื้อ
df1_3
```

```python
df_merge = pd.concat([df1_1,df1_2,df1_3],ignore_index=True) ## รวมตาราง df1_1 df1_2 df1_3
df_merge
```

```python
df_1 = df_merge['sex'].replace({  ## แก้ไขค่าในตารางรวม df_merge
        'ชาย':20,    ## ให้ชาย เป็น 20 
        'หญิง':60   ## ให้หญิง เป็น 60 
    },inplace = True)   ##  แทนที่ในตาราง
```

```python
## Plot ชาย
plt.subplot(1,2,1)
pd.plotting.parallel_coordinates(df_merge_M, 'risk',color=('#556270', '#4ECDC4', '#C7F464'))


## Plot หญิง
plt.subplot(1,2,2)
pd.plotting.parallel_coordinates(df_merge_W, 'risk',color=('#556270', '#4ECDC4', '#C7F464'))
```
